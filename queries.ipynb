```python
from rdflib import Graph, Dataset, RDF
dataset = Dataset()
try:
    dataset.parse("poke-a.nq")
except Exception as e:
    print(e)
```

```python
from collections import Counter
```

```python
def getTypes():
    query = """
    SELECT ?object
    WHERE {
        GRAPH <https://pokemonkg.org/dataset/bulbapedia> {
            ?subject ?predicate ?object .
            FILTER (?predicate = <https://pokemonkg.org/ontology#hasType>)
        }
    }
    """

    results = dataset.query(query)
    types = []
    for row in results:
        types.append(str(row).split("Type:")[1].split("'")[0].lower())

    return types

Counter(getTypes())
```

```python
def getHeights():
    query = """
    SELECT ?object
    WHERE {
        GRAPH <https://pokemonkg.org/dataset/bulbapedia> {
            ?subject ?predicate ?object .
            FILTER (?predicate = <http://qudt.org/schema/qudt/quantityValue>)
            FILTER contains(str(?object),"height")
        }
    }
    """

    results = dataset.query(query)
    heights = []
    for row in results:
        heights.append(str(row).split("value/")[1].split("'")[0].lower())

    return heights

heights = Counter(getHeights())
print(heights)
print(max(heights.keys()))
print(min(heights.keys()))
```
