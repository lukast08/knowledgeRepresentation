```python
def getAllPokemons():
    query = """
    SELECT ?object
    WHERE {
        GRAPH <https://pokemonkg.org/dataset/pokewiki-de-pokemonlist> {
            ?subject ?predicate ?object .
            FILTER (?predicate = <http://www.w3.org/2000/01/rdf-schema#label>)
            FILTER (lang(?object) = "" || lang(?object) = "en")
        }
    }
    """

    results = dataset.query(query)
    pokemons = []
    for result in results:
        pokemon = str(result).split("Literal('")[1].split("'")[0]
        pokemons.append(pokemon)

    return pokemons
```

```python
def getTypes():
    query = """
    SELECT ?object
    WHERE {
        GRAPH <https://pokemonkg.org/dataset/bulbapedia> {
            ?subject ?predicate ?object .
            FILTER (?predicate = <https://pokemonkg.org/ontology#hasType>)
        }
    }
    """

    results = dataset.query(query)
    types = []
    for row in results:
        types.append(str(row).split("Type:")[1].split("'")[0].lower())

    return types

Counter(getTypes())
```

```python
def getHeights():
    query = """
    SELECT ?object
    WHERE {
        GRAPH <https://pokemonkg.org/dataset/bulbapedia> {
            ?subject ?predicate ?object .
            FILTER (?predicate = <http://qudt.org/schema/qudt/quantityValue>)
            FILTER contains(str(?object),"height")
        }
    }
    """

    results = dataset.query(query)
    heights = []
    for row in results:
        heights.append(str(row).split("value/")[1].split("'")[0].lower())

    return heights

heights = Counter(getHeights())
print(heights)
print(max(heights.keys()))
print(min(heights.keys()))
```

```python
def getRegions():
    query = """
    SELECT ?subject
    WHERE {
        GRAPH <http://pokemon.outofbits.com/dataset/pokeapi-co> {
            ?subject ?predicate ?object .
            FILTER (?predicate = <https://pokemonkg.org/ontology#describesPokmon>)
        }
    }
    """

    results = dataset.query(query)
    regions = []
    for row in results:
        regions.append(str(row).split("pokedex/")[1].split("/")[0].lower())

    return regions

Counter(getRegions())
```
